#### BBC covid -vaccine news (Centerist)

```python
import urllib.request,sys,time
from bs4 import BeautifulSoup
import requests
import pandas as pd

#to make the code work through pages
pagesToGet= 1

upperframe=[]  
for page in range(1,pagesToGet+10):
    print('processing page :', page)
    url = 'https://www.bbc.co.uk/search?q=vaccine&page='+str(page)
    print(url)
    
    #an exception might be thrown, so the code should be in a try-except block
    try:
        #use the browser to get the url. This is suspicious command that might blow up.
        page=requests.get(url)                             # this might throw an exception if something goes wrong.
    
    except Exception as e:                                   # this describes what to do if an exception is thrown
        error_type, error_obj, error_info = sys.exc_info()      # get the exception information
        print ('ERROR FOR LINK:',url)                          #print the link that cause the problem
        print (error_type, 'Line:', error_info.tb_lineno)     #print error info and line that threw the exception
        continue                                              #ignore this page. Abandon this and go back.
    time.sleep(2)   
    soup=BeautifulSoup(page.text,'html.parser')
    frame=[]
    links=soup.find_all('div',attrs={'class':'ssrcss-11rb3jo-Promo ett16tt0'})
    print(len(links))
    
    for j in links:
        Headline = j.find("p",attrs={'class':'ssrcss-6arcww-PromoHeadline e1f5wbog4'}).text.strip()
        Description = j.find("p",attrs={'class':'ssrcss-1q0x1qg-Paragraph eq5iqo00'}).text.strip()
    
        frame.append((Headline,Description))
        
    upperframe.extend(frame)

data_bbc=pd.DataFrame(upperframe, columns=['Headline', 'Description'])
data_bbc["Article"] = data_bbc["Headline"].map(str) +" "+ data_bbc["Description"].map(str)

data_bbc.head()
```

### Sentiment Analysis

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyser = SentimentIntensityAnalyzer()

sentiment_results = []
sent_res_neg = []
sent_res_pos = []
sent_res_neu = []
for item in result['Article']:
    sentiment_statement = analyser.polarity_scores(item)
    sentiment_results.append(sentiment_statement['compound'])
    sent_res_neg.append(sentiment_statement['neg'])
    sent_res_pos.append(sentiment_statement['neu'])
    sent_res_neu.append(sentiment_statement['pos'])
```

```python
sentiment_results
```

```python
result["Sentiment Compound Score"] = sentiment_results
result["Negative Sentiment Score"] = sent_res_neg
result["Neutral Sentiment Score"] = sent_res_neu
result["Positive Sentiment Score"] = sent_res_pos
```

```python
result
```

```python
# Create new column to label the sentiment as positive or negative
pos=0
neg=0
neu=0
sntmnt =[]
for i in result['Sentiment Compound Score']:
    if i > 0.2:
        pos+=1
        sntmnt.append('Positive')
    elif i < -0.2:
        sntmnt.append('Negative')
        neg+=1
    else:
        sntmnt.append('Neutral')
        neu+=1
```

```python
# create a dataframe series 
result["Sentiment"] = sntmnt
```
